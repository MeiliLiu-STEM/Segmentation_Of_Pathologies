```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
acc_train_vgg = np.loadtxt('E:/results2017/acc.txt')
acc_train_incep = np.loadtxt('E:/results2017/accu_train_incep1.txt')
```

```python
plt.figure(figsize=(10,5))
line,=plt.plot(acc_train_vgg, linestyle='-.', color='b', label='Training Acc VGG16.')
line.set_dashes([10, 4])
line,=plt.plot(acc_train_incep, linestyle='-', color='r', label='Training Acc Inception.')
plt.xticks(np.arange(0, 110, step=10))
plt.yticks(np.arange(0.5, 1.05, step=0.05))
plt.legend(loc = "lower right")
plt.grid(linewidth=0.3)
plt.title('Comparison Between Models')
plt.show()
```

```python
acc_test_vgg = np.loadtxt('E:/results2017/val_acc.txt')
acc_test_incep = np.loadtxt('E:/results2017/accu_test_incep1.txt')
```

```python
plt.figure(figsize=(10,5))
line,=plt.plot(acc_test_vgg, linestyle='-', color='b', label='Test Acc VGG16.')
plt.plot(acc_test_incep, linestyle='-', color='r', label='Test Acc Inception.')
plt.xticks(np.arange(0, 110, step=10))
plt.yticks(np.arange(0.5, 1.05, step=0.05))
plt.legend()
plt.grid(linewidth=0.3)
plt.title('Comparison Between Models')
plt.show()
```
