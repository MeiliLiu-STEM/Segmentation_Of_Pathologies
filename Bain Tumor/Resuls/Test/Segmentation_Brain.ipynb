```python
def transpose_volumes(volumes, slice_direction):
    if (slice_direction == 'axial'):
        tr_volumes = volumes
    elif(slice_direction == 'sagittal'):
        tr_volumes = np.transpose(volumes.get_data(), (2, 0, 1))
    elif(slice_direction == 'coronal'):
        tr_volumes = np.transpose(volumes.get_data(), (1, 2, 0))
    else:
        print('undefined slice direction:', slice_direction)
        tr_volumes = volumes
    return tr_volumes
```

```python
def get_patient_data(fileName,direction):
    image =nibabel.load(fileName)
    image_ax = transpose_volumes(image,direction)
    if(direction == 'axial'):
        data_ax = image_ax.get_data()
    else:
        data_ax = image_ax
    return np.asarray(data_ax)
```
