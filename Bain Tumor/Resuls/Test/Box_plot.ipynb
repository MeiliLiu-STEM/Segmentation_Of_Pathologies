```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/Nassim/Downloads/Compressed/Results/Stats_Validation.csv', delimiter=',')
with open('D:/Last_result/resultts2017/dice_core.txt') as f:
    content = f.readlines()
dice_core = [float(x.strip()) for x in content]
with open('D:/Last_result/resultts2017/dice_enhac.txt') as f:
    content = f.readlines()
dice_enhac = [float(x.strip()) for x in content]
with open('D:/Last_result/resultts2017/dice_whole.txt') as f:
    content = f.readlines()
dice_whole = [float(x.strip()) for x in content]
```

```python
all_data = [df['Dice_ET'][:46],df['Dice_WT'][:46],df['Dice_TC'][:46]]
all_data1 = [dice_enhac,dice_whole,dice_core]
labels = ['Dice ET','Dice WT','Dice TC']
fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(121)
bplot = ax.boxplot(all_data1,
                         vert=True,  # vertical box alignment
                         patch_artist=True,  # fill with color
                         labels=labels,showmeans=True,showfliers = False)  # will be used to label x-ticks
ax.set_title('Test dataset Box plot',fontsize=16)

i=0
colors = ['#823939', '#d78658','#c3c3c3']
for box in bplot['boxes']:
    box.set( color=colors[i], linewidth=2)
    box.set( facecolor = colors[i])
    i+=1
for median in bplot['medians']:
    median.set(color='black', linewidth=2)
for whisker in bplot['whiskers']:
    whisker.set(color='#7570b3', linewidth=2, linestyle = '--')
for mean in bplot['means']:
    mean.set(linewidth=4,marker = 's',color = 'r')
ax.yaxis.grid(True)
ax.set_ylabel('Dice Score',labelpad = 10,size = 16)
ax.set_xlabel('(A)',labelpad = 10,size = 16)
from matplotlib import rcParams
rcParams['xtick.labelsize'] = 16
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
ax = fig.add_subplot(122)
bplot = ax.boxplot(all_data,
                         vert=True,  # vertical box alignment
                         patch_artist=True,  # fill with color
                         labels=labels,showmeans=True,showfliers = False)  # will be used to label x-ticks
ax.set_title('Validation dataset Box plot',fontsize=16)

i=0
colors = ['#823939', '#d78658','#c3c3c3']
for box in bplot['boxes']:
    box.set( color=colors[i], linewidth=2)
    box.set( facecolor = colors[i])
    i+=1
for median in bplot['medians']:
    median.set(color='black', linewidth=2)
for whisker in bplot['whiskers']:
    whisker.set(color='#7570b3', linewidth=2, linestyle = '--')
for mean in bplot['means']:
    mean.set(linewidth=4,marker = 's',color = 'r')
ax.yaxis.grid(True)
ax.set_ylabel('Dice Score',labelpad = 10,size = 16)
ax.set_xlabel('(B)',labelpad = 10,size = 16)
from matplotlib import rcParams
rcParams['xtick.labelsize'] = 16
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
plt.show()
```
